## Segment 4 - Concatenating and transforming data

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

```python
DF_obj = pd.DataFrame(np.arange(36).reshape(6,6)) # Generate value series from 0 to 35 in 6x6 table
DF_obj
```

```python
DF_obj_2 = pd.DataFrame(np.arange(15).reshape(5,3)) # Value series 0 to 14 in 5x3 table
DF_obj_2
```

### Concatenating data

```python
# Concat method = join data from separate srounces into 1 combined table
# axis = 1 argument concats dataframes based on row index
pd.concat([DF_obj, DF_obj_2], axis=1)
```

```python
# No axis=1 ---> concatenated on column index
pd.concat([DF_obj, DF_obj_2])
```

### Transforming data
#### Dropping data

```python
DF_obj.drop([0,2]) # drop rows 0 and 2
```

```python
# axis=1 drops columns instead of rows
DF_obj.drop([0,2], axis = 1)
```

### Adding data

```python
series_obj = Series(np.arange(6)) # Create series of 0 to 5
series_obj.name = "added_variable"
series_obj
```

```python
variable_added = DataFrame.join(DF_obj, series_obj) # Join dataframe and series object
variable_added
```

```python
# append variable_added datatable to itself
added_datatable = variable_added.append(variable_added, ignore_index=False) # last bit leaves index values as the are (doesn't reindex)
added_datatable
```

```python
added_datatable = variable_added.append(variable_added, ignore_index=True) # regenerate index values accurate to table by removing ignore_index=False
added_datatable
```

### Sorting data

```python
# Sort data by items in column 5 and in descending order
DF_sorted = DF_obj.sort_values(by=5, ascending=[False])
DF_sorted
```
